```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn = sqlite3.connect("food_delivery.db")
cursor=conn.cursor()
```

```python
with open("schema.sql","r") as f:
    cursor.executescript(f.read())
with open("data.sql","r") as f:
    cursor.executescript(f.read())
    
conn.commit()
```

Query 1: Average Delivery Time by Restaurant

```python
query= """ SELECT r.name AS restaurant,
       ROUND(AVG((julianday(d.dropoff_time) - julianday(d.pickup_time)) * 24 * 60), 2) AS avg_delivery_minutes
FROM deliveries d
JOIN orders o ON d.order_id = o.order_id
JOIN restaurants r ON o.restaurant_id = r.restaurant_id
GROUP BY r.name
ORDER BY avg_delivery_minutes DESC; """

df = pd.read_sql_query(query, conn)
sns.barplot(data=df, x='avg_delivery_minutes', y='restaurant')
plt.title("Average Delivery Time by Restaurant")
plt.xlabel("Minutes")
plt.ylabel("Restaurant")
plt.tight_layout()
plt.show()
```

Query 2: Monthly Order Volume

```python
query = """
SELECT strftime('%Y-%m', order_time) AS month,
       COUNT(*) AS total_orders
FROM orders
GROUP BY month
ORDER BY month;
"""
df = pd.read_sql_query(query, conn)
df.plot(x="month", y="total_orders", kind="line", marker="o", title="Monthly Order Volume")
plt.xticks(rotation=45)
plt.ylabel("Orders")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Query 3: Peak Ordering Hours

```python
query = """
SELECT strftime('%H', order_time) AS hour,
       COUNT(*) AS order_count
FROM orders
GROUP BY hour
ORDER BY order_count DESC
LIMIT 5;
"""
df = pd.read_sql_query(query, conn)
sns.barplot(data=df, x="hour", y="order_count")
plt.title("Top 5 Peak Ordering Hours")
plt.ylabel("Orders")
plt.xlabel("Hour of Day")
plt.tight_layout()
plt.show()
```

Customer Retention Table

```python
query = """
SELECT customer_id,
       COUNT(order_id) AS order_count,
       MIN(order_time) AS first_order,
       MAX(order_time) AS last_order
FROM orders
GROUP BY customer_id
HAVING COUNT(order_id) > 1;
"""
df = pd.read_sql_query(query, conn)
display(df)
```

Regional Weekly Demand

```python
query = """
SELECT region,
       strftime('%Y-%W', order_time) AS week,
       COUNT(*) AS orders
FROM orders
GROUP BY region, week
ORDER BY orders DESC
LIMIT 10;
"""
df = pd.read_sql_query(query, conn)
sns.barplot(data=df, x="orders", y="region", hue="week")
plt.title("Regional Demand Spikes (Top 10 Weeks)")
plt.tight_layout()
plt.show()
```
